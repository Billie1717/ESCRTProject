# Experiments

```python
# new legend

clrs = ["#949494", "#2a9359", "#e09638" ] # Claudia's gray - green -yellow
clrs = ["#eda385","#51c18c","#f2d664"]   # Claudia's orange - green -yellow
fig1, ax1 = plt.subplots(1, figsize = (5,4))
plt.fill_between(LB0x, LB0, UB0[0:69], color = clrs[0], alpha=0.2, lw=0, edgecolor="white")
plt.fill_between(LB2x, LB2, UB2[5:69], color =  clrs[1], alpha=0.2, lw=0, edgecolor='white')
plt.fill_between(UB3x, LB3[10:64], UB3, color =  clrs[2], alpha=0.2, lw=0, edgecolor='white')
plt.gcf().subplots_adjust(left=0.18)
plt.gcf().subplots_adjust(right=0.8)
plt.gcf().subplots_adjust(bottom=0.18)
plt.yticks([0,1],["0","1"], size = 17)

ax1.set_xlim((-1,29))
ax1.set_ylim((-0.2,1.05))
ax1.get_xaxis().set_ticks([])

ax1.plot(x0,P0,color =  clrs[0],linewidth=2.0, label = "Monomer 0 (Snf7)")
ax1.plot(x2,P2,color = clrs[1],linewidth=2.0,label = "Monomer 1 (Vps2)")
ax1.plot(x3,P3,color = clrs[2],linewidth=2.0,label = "Monomer 2 (Did2)")

    
ax1.set_xlabel(r"time", size = 17, labelpad=10)
ax1.set_ylabel("Relative fluorescence", size = 17, loc='bottom')


plt.yticks([0,1],["0","1"], size = 17)
lgd = ax1.legend(fontsize=14, loc = (0.205,0),frameon=False, ncol = 1,labelspacing=0.3, handlelength=1.5, handletextpad=1)

plt.show()

fname  = 'Experimental.pdf'
fig1.savefig('Plots/'+fname)
```

# Plotting individual prob curves sequence

```python
loc = datadir+'/ModelData/'
P0 = pandas.read_csv(loc + "Arrows_dProb0.txt", header = None)[0]
P1 = pandas.read_csv(loc + "Arrows_dProb1.txt", header = None,  usecols= [0])[0]
P2 = pandas.read_csv(loc + "Arrows_dProb2.txt", header = None, usecols= [0])[0]
#loc = datadir+'DataDoS/'
#P0 = pandas.read_csv(loc + "smolspaceTest1_Prob0.txt", header = None)[0]
#P1 = pandas.read_csv(loc + "smolspaceTest1_Prob1.txt", header = None,  usecols= [0])[0]
#P2 = pandas.read_csv(loc + "smolspaceTest1_Prob2.txt", header = None, usecols= [0])[0]
N=42
fig1, ax1 = plt.subplots(1, figsize = (5,4))
ax1.plot(100*P0[0:5000]/N,color = "#53a8c9",linewidth=3.0, label = "Monomer 0")
ax1.plot(100*P1[0:5000]/N,color = "#7e6ca8",linewidth=3.0,label = "Monomer 1")
ax1.plot(100*P2[0:5000]/N,color = "#dd6fb6",linewidth=3.0,label = "Monomer 2")
ax1.set_xlabel(r"time", size = 17)
ax1.set_ylabel("% lattice coverage", size = 17)
#ax1.legend(frameon=False)

n = 2  # Keeps every 2nd label
[l.set_visible(False) for (i,l) in enumerate(ax1.xaxis.get_ticklabels()) if i % n != 0]
[l.set_visible(False) for (i,l) in enumerate(ax1.yaxis.get_ticklabels()) if i % n != 0]

textstr = '\n'.join((
    r'$p_{2}>p_{1}>p_{0}$' ,
    r'$c_{2}>c_{1}>c_{0}$' ,
    r'$E_{2}^\mathrm{adh}>E_{1}^\mathrm{adh}>E_{0}^\mathrm{adh}$' ))

ax1.text(0.02, 0.97, textstr, transform=ax1.transAxes, fontsize=16,verticalalignment='top')

plt.xticks([0,1000,2000,3000,4000,5000],["0","1","2","3","4","5"], size = 15)
plt.yticks([0,25,50,75,100],size = 15)
plt.gcf().subplots_adjust(left=0.18)
plt.gcf().subplots_adjust(bottom=0.18)
plt.show()

#fname  = 'Test1.pdf'
fname  = 'Arrows_d.pdf'
fig1.savefig('Plots/'+fname)
```

# Plotting curves: Identical filaments

```python
# Smaller plot

loc = datadir+'/ModelData/'
P0 = pandas.read_csv(loc + "Arrows_aProb0.txt", header = None)[0]
P1 = pandas.read_csv(loc + "Arrows_aProb1.txt", header = None,  usecols= [0])[0]
P2 = pandas.read_csv(loc + "Arrows_aProb2.txt", header = None, usecols= [0])[0]

N=42
fig1, ax1 = plt.subplots(1, figsize = (3*1.25,3))

ax1.plot(100*P0/N,color = "#53a8c9",linewidth=3.0, label = "Monomer 0")
ax1.plot(100*P1/N,color = "#7e6ca8",linewidth=3.0,label = "Monomer 1")
ax1.plot(100*P2/N,color = "#dd6fb6",linewidth=3.0,label = "Monomer 2")
    
ax1.set_xlabel(r"time", size = 17)
ax1.set_ylabel("% lattice coverage", size = 17)
#ax1.legend(frameon=False)

n = 2  # Keeps every 2nd label
[l.set_visible(False) for (i,l) in enumerate(ax1.xaxis.get_ticklabels()) if i % n != 0]
n=2
[l.set_visible(False) for (i,l) in enumerate(ax1.yaxis.get_ticklabels()) if i % n != 0]

textstr = '\n'.join((
    r'$p_{2}=p_{1}=p_{0}$' ,
    r'$c_{2}=c_{1}=c_{0}$' ,
    r'$E_{2}^{\mathrm{adh}}=E_{1}^{\mathrm{adh}}=E_{0}^{\mathrm{adh}}$' ))

ax1.text(0.25, 0.5, textstr, transform=ax1.transAxes, fontsize=16,verticalalignment='top')


plt.xticks([0,1000,2000,3000,4000],["0","1","2","3","4"], size = 15)
plt.yticks([0,25,50,75],size = 15)

plt.gcf().subplots_adjust(left=0.15)
plt.gcf().subplots_adjust(bottom=0.18)
plt.show()

fname  = 'Arrows_a.pdf'
fig1.savefig('Plots/'+fname)
```
