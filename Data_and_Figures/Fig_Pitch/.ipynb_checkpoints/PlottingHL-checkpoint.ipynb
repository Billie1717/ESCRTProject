```python
import matplotlib.pyplot as plt
import pandas
import numpy as np


datadir = '/Users/billiemeadowcroft/Dropbox/PhDGithub/RemoteESCRT/Data_and_Figures/Fig_Pitch/'
loc2 = datadir+'HalfLifeSameBind'
P1s = 20
P2s = 1

HL0 = pandas.read_csv(loc2 + ".txt", header = None, sep=',', usecols= [10])
HL1 = pandas.read_csv(loc2 + ".txt", header = None, sep=',', usecols= [11])
HL2 = pandas.read_csv(loc2 + ".txt", header = None, sep=',', usecols= [12])

HL0_arr = HL0.values.reshape(P1s,P2s)
HL1_arr = HL1.values.reshape(P1s,P2s)
HL2_arr = HL2.values.reshape(P1s,P2s)
```

```python



fig1, axs1 = plt.subplots(1,figsize=(6,5))
fig1.tight_layout()
x = np.arange(0, 20, 1)*(150/20)  #sub2
X=np.round(x,0)

plt.rc('xtick', labelsize=15) 
plt.rc('ytick', labelsize=15) 

axs1.set_ylabel(r'Time to half max, $t_{1/2}$ [$\Delta t$]', fontsize = 20)
axs1.set_xlabel(r'$p_1$ [nm]', size = 20)
plt.xticks(size = 13)
plt.yticks(size = 15)

axs1.plot(X,HL0_arr,color = "skyblue",linewidth=3.0, label = "Monomer 0")
axs1.plot(X,(HL1_arr),color = "mediumpurple",linewidth=3.0, label = "Monomer 1")
axs1.plot(X,(HL2_arr),color = "plum",linewidth=3.0, label = "Monomer 2")

plt.gcf().subplots_adjust(left=0.15)
plt.gcf().subplots_adjust(bottom=0.15)
plt.gcf().subplots_adjust(top=0.8)
axs1.legend(frameon=False, loc = (0.1,0.65),prop={'size': 17})
#plt.show()
plt.xticks(size = 13)
plt.yticks(size = 15)

fname  = 'HalfLifeLineplots.pdf'
fig1.savefig('Plots/'+fname)


```
