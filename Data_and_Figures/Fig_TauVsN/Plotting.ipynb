## Code for looking at binding VS membrane Def for multiple filaments

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib  import cm
from matplotlib.colors import LogNorm, Normalize
```

## Plotting minimum binding energy Vs number of filaments

```python
fig2,ax2=plt.subplots()

FilBar = [281, 44.6,5.6,0,0,0,0]
Nsq = [1,4,9,16,25,36,49]

ax2.plot([1,2,3,4,5,6,7],FilBar+np.log(Nsq), marker = 'o',markersize = '8',color = 'mediumpurple')
ax2.set_yscale('log')
ax2.set_xticks([1,2,3,4,5,6,7])
ax2.set_yticks([280,40,5,2])
ax2.set_yticklabels(['280','40','5','2'])
ax2.set_xticklabels(['One','Two','Three','Four','Five','Six','Seven'])
ax2.set_xlabel('Number of filament types ($N_{\mathrm{types}}$)', fontsize = 12)
ax2.set_ylabel('Optimality\n'+r'$ln(\tau)$, $\tau \sim N_{\mathrm{types}}^{2}e^{\Delta \ E}$', fontsize = 15)
textstr = '\n'.join((
    r'$\bf{Last \ Filament:}$' ,
     r'Pitch = 75nm' ,
    r'Radius = 6.25nm',
    r'$E_{\mathrm{adh}} = 29.8 k_{\mathrm{B}}T$'))
ax2.text(0.4, 0.7, textstr, transform=ax2.transAxes, fontsize=15,verticalalignment='top')

plt.gcf().subplots_adjust(left=0.15)

fig2.tight_layout()

fname = 'LogTau1.pdf'
fig2.savefig('./Plots/'+fname)
#fname2 = 'SameBarriersVSfilNumberExpRescaled.pdf'
#fig3.savefig('./Plots/'+fname2)
```
